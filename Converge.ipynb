```
import math
import matplotlib.pyplot as plt


def g1(inicial, criterio_paro = 0.0001):
    
    x = [inicial]
    i=0
    x.append((2*(x[i]**2))-5)
    
    i+=1
    error_vec=[1]
    error_vec.append(abs(x[i] - x[i-1]))
    confirmar = True
    
    Condicion_Convergir=0
    confirmar_mensaje='Sí converge'
    while (x[i]<90000000 and (error_vec[i-1] > criterio_paro)):
            
            x.append((2*(x[i]**2))-5)
            Condicion_Convergir+=1
            i+=1
            error_vec.append(abs(x[i] - x[i-1]))
            if(x[i]>90000000 and (error_vec[i-1] > criterio_paro)):
                confirmar = False
                confirmar_mensaje='No converge'
            
            
    print('\ng1')
    print('x: '+ str(x) )
    print('\nerror_vec: ' + str(error_vec))
    plt.plot(x)
    plt.title("g1 x")
    plt.show()
    plt.plot(error_vec)
    plt.title("g1 error_vec")
    plt.show()
    
    return str(confirmar)+'  ' + confirmar_mensaje

def g2(inicial, criterio_paro = 0.0001):
    
    x = [inicial]
    i=0
    x.append(5/(2*x[i]-1))
    
    i+=1
    error_vec=[1]
    error_vec.append(abs(x[i] - x[i-1]))
    confirmar = True
    contador = 0
    Condicion_Convergir=0
    confirmar_mensaje='Sí converge'
    while (x[i]<90000000 and (error_vec[i-1] > criterio_paro) and contador<60):
            
            x.append(5/(2*x[i]-1))
            Condicion_Convergir+=1
            i+=1
            error_vec.append(abs(x[i] - x[i-1]))
            contador+=contador
            
            if(x[i]>90000000 and contador<60):
                
                confirmar = False
                confirmar_mensaje='No converge'
            
    print('\ng2')
    print('x: '+ str(x) )
    print('\nerror_vec: ' + str(error_vec))
    plt.plot(x)
    plt.title("g2 x")
    plt.show()
    plt.plot(error_vec)
    plt.title("g2 error_vec")
    plt.show()
    
    return str(confirmar)+'  ' + confirmar_mensaje


def g3(inicial, criterio_paro = 0.0001):
    
    x = [inicial]
    i=0
    x.append(-1*(math.sqrt((x[i]+5)/2)))
    i+=1
    error_vec=[1]
    error_vec.append(abs(x[i] - x[i-1]))
    confirmar = True
    contador = 0
    Condicion_Convergir=0
    confirmar_mensaje='Sí converge'
    while (x[i]<90000000 and (error_vec[i-1] > criterio_paro) and contador<60):
            
            x.append(-1*(math.sqrt((x[i]+5)/2)))
            Condicion_Convergir+=1
            i+=1
            error_vec.append(abs(x[i] - x[i-1]))
            contador+=contador
            
            if(x[i]>90000000 and contador<60):
                
                confirmar = False
                confirmar_mensaje='No converge'
            
            
            
    print('\ng3 Negativo')
    print('x: '+ str(x) )
    print('\nerror_vec: ' + str(error_vec))
    plt.plot(x)
    plt.title("g3 x Negativo")
    plt.show()
    plt.plot(error_vec)
    plt.title("g3 error_vec Negativo")
    plt.show()
    
    
    x2 = [inicial]
    i=0
    x2.append(math.sqrt((x2[i]+5)/2))
    i+=1
    error_vec2=[1]
    error_vec2.append(abs(x2[i] - x2[i-1]))
    confirmar2 = True
    contador = 0
    Condicion_Convergir=0
    confirmar_mensaje2='Sí converge'
    while (x[i]<90000000 and (error_vec2[i-1] > criterio_paro) and contador<60):
            
            x2.append(math.sqrt((x2[i]+5)/2))
            Condicion_Convergir+=1
            i+=1
            error_vec2.append(abs(x[i] - x[i-1]))
            contador+=contador
            
            if(x[i]>90000000 and contador<60):
                
                confirmar2 = False
                confirmar_mensaje2='No converge'
            
            
            
    print('\ng3 Positivo')
    print('x: '+ str(x) )
    print('\nerror_vec: ' + str(error_vec))
    plt.plot(x)
    plt.title("g3 x Positivo")
    plt.show()
    plt.plot(error_vec)
    plt.title("g3 error_vec Positivo")
    plt.show()
    
    return str(confirmar)+'  ' + confirmar_mensaje   +'  ' +str(confirmar2)+'  ' + confirmar_mensaje2
    


x = int(input('Digite su valor incial \n'))
automatico=int(input("Presione el numero 0 para tener criterio de paro automataico o cualquiero otro numero para ingresar uno   \n"))
if automatico==0:
    print(str(g1(x)))
    print(str(g2(x)))
    print(str(g3(x)))
else:  
    criterio_paro = float(input('Digite su valor del criterio de paro \n'))
    print(str(g1(x,criterio_paro)))
    print(str(g2(x,criterio_paro)))
    print(str(g3(x,criterio_paro)))
```
